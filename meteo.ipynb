```python
# Import library
import glob
import pandas  as pd #Data manipulation
import numpy as np #Data manipulation
import matplotlib.pyplot as plt # Visualization
import seaborn as sns #Visualization
```

```python
meteo_data = []
for infile in glob.glob("*.csv"):
    data = pd.read_csv(infile)
    # store DataFrame in list
    meteo_data.append(data)
# see pd.concat documentation for more info
meteo_data = pd.concat(meteo_data)
meteo_data.info()
```

```python
meteo_data["DATE"] = pd.to_datetime(meteo_data["DATE"], infer_datetime_format=True)
```

```python
meteo_data["LIEU"].value_counts()
```

```python
#pip install sqlalchemy 
#importing sql library
from sqlalchemy import create_engine
# create a referrence 
# for sql library
engine = create_engine('sqlite://', echo = False)
```

```python
meteo = meteo_data.copy()
```

```python
# attach the data frame to the sql 
# with a name of the table 
# as "meteo"
meteo.to_sql('meteo', con = engine)
```

```python
db = pd.read_sql('SELECT * FROM meteo', engine)
```

```python
meteo_angers = pd.read_sql("SELECT * FROM meteo where LIEU = 'Angers'", engine)
```
